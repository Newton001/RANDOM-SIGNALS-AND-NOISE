```python
#Libraries Import
import numpy as np
import scipy
import matplotlib.pyplot as plt
from scipy import signal


# Seed the random number generator
np.random.seed(1)

# Generate random numbers (100000)
vector1 = np.random.uniform(1, 10, size=10000)
vector2 = np.random.uniform(2, 10, size=10000)

print("Vector 1:", vector1)
print("Vector 2:", vector2)
```

```python
#Computing Pearson's r
r = scipy.stats.pearsonr(vector1, vector2)[0]    

#Computing Spearman's rho
rho = scipy.stats.spearmanr(vector1, vector2)[0]

#Computing Kendall's tau
tau = scipy.stats.kendalltau(vector1, vector2)[0]


print("Pearson's r: ", r)
print("Spearman's rho", rho)
print("Kendall's tau", tau)
```

```python
#Generating another set of Vectors
X = np.random.randint(0,high=10000, size=10000)
Y = np.random.randint(0,high=10000, size=10000)
print("Vector 1: ", X, "Length is:",len(X))
print("Vector 2: ", Y, "Length is: ",len(Y))

#Find Number of points in each vector
num_pointsX = np.size(X)
num_pointsY = np.size(Y)

#Vector Means
X_mean = np.mean(X)
Y_mean = np.mean(Y)

#Calculating Cross deviations and deviations about X
SS_XY = np.sum(X*Y) - num_pointsX*X_mean*Y_mean
SS_XX = np.sum(X*Y) - num_pointsX*X_mean*X_mean

#Regression coefficients
b_1 = SS_XY/SS_XX
b_0 = Y_mean - b_1*X_mean
print("Estimated Coefficients: ")
print("Y intercept is: ", format(b_0,"0.6f"))
print("Slope is = ", format(b_1,"0.6f"))

#Plotting the two vectors in 2D
plt.scatter(X,Y, color = "g", marker= "o", s=10)

#Predicted Response
y_pred = b_0 +b_1*X

#Plotting the Regression line
plt.plot(X, y_pred, color ="r")

#Labels
plt.xlabel("Vector 1")
plt.ylabel("Vector 2")

#Display
plt.show()
```

```python
#Generating another set of Vectors
X1 = np.random.randint(0,high=10000, size=10000)
Y1 = np.random.randint(0,high=10000, size=10000)

#Determine Lengths of Vectors
lgth_X1 = len(X1)
lgth_Y1 = len(Y1)

#Finding Rank for X1

X1_array = np.array(X1)

temp = X1_array.argsort()

RankX1 = np.empty_like(temp)

RankX1[temp] = np.arange(len(X1_array))

#Finding Rank for Y1
Y1_array = np.array(Y1)

temp1 = Y1_array.argsort()

RankY1 = np.empty_like(temp1)

RankY1[temp1] = np.arange(len(Y1_array))

      
print("Vector 1: ", X1, "Length is:",lgth_X1)
print("Vector 2: ", Y1, "Length is: ",lgth_Y1)


#Find Number of points in each vector Rank
num_pointsX1 = np.size(RankX1)
num_pointsY1 = np.size(RankY1)


print(RankY1)
#Vector Means
X1_mean = np.mean(RankX1)
Y1_mean = np.mean(RankY1)

#Calculating Cross deviations and deviations about X
SS_X1Y1 = np.sum(RankX1*RankY1) - num_pointsX1*X1_mean*Y1_mean
SS_X1X1 = np.sum(RankX1*RankY1) - num_pointsX1*X1_mean*X1_mean

#Regression coefficients
b_2 = SS_X1Y1/SS_X1X1
b_3 = Y1_mean - b_2*X1_mean
print("Estimated Coefficients: ")
print("Y intercept is: ", format(b_3,"0.6f"))
print("Slope is = ", format(b_2,"0.6f"))

#Plotting the two vectors in 2D
plt.scatter(X1,Y1, color = "m", marker= "o", s=10)

#Predicted Response
y1_pred = b_2 +b_3*X

#Plotting the Regression line
plt.plot(X1, y1_pred, color ="b")

#Labels
plt.xlabel("Vector 1")
plt.ylabel("Vector 2")

#Display
plt.show()
```
